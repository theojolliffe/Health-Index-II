```python
areas = topindex['AREANM']
area = areas[0]
years = [2015, 2016, 2017, 2018, 2019]
yearLU = {
    2015: 'time1',
    2016: 'time2',
    2017: 'time3',
    2018: 'time4',
    2019: 'time5',
}
```

```python
domains = list(domain_data['DATANM'].unique())

def get_subdomains(x):
    return list(hier[hier['DomainNM'] == x]['SubdomainNM'].unique())

def get_indicators(x):
    return list(hier[hier['SubdomainNM'] == x]['IndicatorNM'].unique())
```

```python
subdomain_data = subdomain_data.rename(columns={"Time1": "time1", "Time2": "time2", "Time3": "time3", "Time4": "time4", "Time5": "time5"})
domain_data = domain_data.rename(columns={"Time1": "time1", "Time2": "time2", "Time3": "time3", "Time4": "time4", "Time5": "time5"})
topindex = topindex.rename(columns={"Time1": "time1", "Time2": "time2", "Time3": "time3", "Time4": "time4", "Time5": "time5"})
```

```python
for area in areas:
    healthindexdata = {}

    data = {}
    indicator_data = indicator_data[indicator_data['AREACD'].map(lambda x: x[:2]=='E0')]
    subdomain_data = subdomain_data[subdomain_data['AREACD'].map(lambda x: x[:2]=='E0')]
    domain_data = domain_data[domain_data['AREACD'].map(lambda x: x[:2]=='E0')]
    for domain in domains:
        domain_filt  = domain_data[(domain_data['DATANM'] == domain)]

        data[domain] = {}
        data[domain]['subdomains'] = {}
        for subdomain in get_subdomains(domain):
            subdomain_filt  = subdomain_data[(subdomain_data['DATANM'] == subdomain)]

            data[domain]['subdomains'][subdomain] = {}
            data[domain]['subdomains'][subdomain]['indicators'] = {}

            for indicator in get_indicators(subdomain):
                indicator_filt  = indicator_data[(indicator_data['DATANM'] == indicator)]

                for year in years:
                    data[domain]['subdomains'][subdomain]['indicators'][indicator] = {}
                    data[domain]['subdomains'][subdomain]['indicators'][indicator][year] = {}

                    # VALUE
                    data[domain]['subdomains'][subdomain]['indicators'][indicator][year]['value'] = round(indicator_filt[indicator_filt['AREANM'] == area].iloc[0][yearLU[year]], 1)

                    # RANK
                    data[domain]['subdomains'][subdomain]['indicators'][indicator][year]['rank'] = list(indicator_filt.sort_values(by=yearLU[year], ascending=False)['AREANM']).index(area)+1

                    # CHANGE1YEAR
                    if year == 2015:
                        data[domain]['subdomains'][subdomain]['indicators'][indicator][year]['Change1year'] = 'NaN'
                        data[domain]['subdomains'][subdomain]['indicators'][indicator][year]['Change1year Rank'] = 'NaN'
                    else:
                        change1 = indicator_filt[indicator_filt['AREANM'] == area].iloc[0][yearLU[year]] - indicator_filt[indicator_filt['AREANM'] == area].iloc[0][yearLU[year-1]]
                        data[domain]['subdomains'][subdomain]['indicators'][indicator][year]['Change1year'] = round(change1, 1)

                        change1list = sorted([indicator_filt.loc[i][yearLU[year]] - indicator_filt.loc[i][yearLU[year-1]] for i in indicator_filt.index], reverse=True)
                        data[domain]['subdomains'][subdomain]['indicators'][indicator][year]['Change1year Rank'] = change1list.index(change1) + 1

                    # CHANGE4YEAR
                    if year == 2019:
                        change4 = indicator_filt[indicator_filt['AREANM'] == area].iloc[0][yearLU[2019]] - indicator_filt[indicator_filt['AREANM'] == area].iloc[0][yearLU[2015]]
                        data[domain]['subdomains'][subdomain]['indicators'][indicator][year]['Change4year'] = round(change4, 1)

                        change4list = sorted([indicator_filt.loc[i][yearLU[2019]] - indicator_filt.loc[i][yearLU[2015]] for i in indicator_filt.index], reverse=True)
                        data[domain]['subdomains'][subdomain]['indicators'][indicator][year]['Change4year Rank'] = change4list.index(change4) + 1
                    else:
                        data[domain]['subdomains'][subdomain]['indicators'][indicator][year]['Change4year'] = 'NaN'
                        data[domain]['subdomains'][subdomain]['indicators'][indicator][year]['Change4year Rank'] = 'NaN'

            data[domain]['subdomains'][subdomain]['total'] = {}
            for year in years:
                data[domain]['subdomains'][subdomain]['total'][year] = {}

                # VALUE
                data[domain]['subdomains'][subdomain]['total'][year]['value'] = round(subdomain_filt[subdomain_filt['AREANM'] == area].iloc[0][yearLU[year]], 1)

                # RANK
                data[domain]['subdomains'][subdomain]['total'][year]['rank'] = list(subdomain_filt.sort_values(by=yearLU[year], ascending=False)['AREANM']).index(area)+1

                # CHANGE1YEAR
                if year == 2015:
                    data[domain]['subdomains'][subdomain]['total'][year]['Change1year'] = 'NaN'
                    data[domain]['subdomains'][subdomain]['total'][year]['Change1year Rank'] = 'NaN'
                else:
                    change1 = subdomain_filt[subdomain_filt['AREANM'] == area].iloc[0][yearLU[year]] - subdomain_filt[subdomain_filt['AREANM'] == area].iloc[0][yearLU[year-1]]
                    data[domain]['subdomains'][subdomain]['total'][year]['Change1year'] = round(change1, 1)

                    change1list = sorted([subdomain_data.loc[i][yearLU[year]] - subdomain_data.loc[i][yearLU[year-1]] for i in subdomain_data.index], reverse=True)
                    data[domain]['subdomains'][subdomain]['total'][year]['Change1year Rank'] = change1list.index(change1) + 1

                # CHANGE4YEAR
                if year == 2019:
                    change4 = subdomain_filt[subdomain_filt['AREANM'] == area].iloc[0][yearLU[2019]] - subdomain_filt[subdomain_filt['AREANM'] == area].iloc[0][yearLU[2015]]
                    data[domain]['subdomains'][subdomain]['total'][year]['Change4year'] = round(change4, 1)

                    change4list = sorted([subdomain_data.loc[i][yearLU[2019]] - subdomain_data.loc[i][yearLU[2015]] for i in subdomain_filt.index], reverse=True)
                    data[domain]['subdomains'][subdomain]['total'][year]['Change4year Rank'] = change4list.index(change4) + 1
                else:
                    data[domain]['subdomains'][subdomain]['total'][year]['Change4year'] = 'NaN'
                    data[domain]['subdomains'][subdomain]['total'][year]['Change4year Rank'] = 'NaN'

        data[domain]['total'] = {}
        for year in years:
            data[domain]['total'][year] = {}

            # VALUE
            data[domain]['total'][year]['value'] = round(domain_filt[domain_filt['AREANM'] == area].iloc[0][yearLU[year]], 1)

            # RANK
            data[domain]['total'][year]['rank'] = list(domain_filt.sort_values(by=yearLU[year], ascending=False)['AREANM']).index(area)+1

            # CHANGE1YEAR
            if year == 2015:
                data[domain]['total'][year]['Change1year'] = 'NaN'
                data[domain]['total'][year]['Change1year Rank'] = 'NaN'
            else:
                change1 = domain_filt[domain_filt['AREANM'] == area].iloc[0][yearLU[year]] - domain_filt[domain_filt['AREANM'] == area].iloc[0][yearLU[year-1]]
                data[domain]['total'][year]['Change1year'] = round(change1, 1)

                change1list = sorted([domain_data.loc[i][yearLU[year]] - domain_data.loc[i][yearLU[year-1]] for i in domain_filt.index], reverse=True)
                data[domain]['total'][year]['Change1year Rank'] = change1list.index(change1) + 1

            # CHANGE4YEAR
            if year == 2019:
                change4 = domain_filt[domain_filt['AREANM'] == area].iloc[0][yearLU[2019]] - domain_filt[domain_filt['AREANM'] == area].iloc[0][yearLU[2015]]
                data[domain]['total'][year]['Change4year'] = round(change4, 1)

                change4list = sorted([domain_data.loc[i][yearLU[2019]] - domain_data.loc[i][yearLU[2015]] for i in domain_filt.index], reverse=True)
                data[domain]['total'][year]['Change4year Rank'] = change4list.index(change4) + 1
            else:
                data[domain]['total'][year]['Change4year'] = 'NaN'
                data[domain]['total'][year]['Change4year Rank'] = 'NaN'

    # OVERALL

    data['Overall'] = {}
    data['Overall']['total'] = {}
    topindex = topindex[topindex['AREACD'].map(lambda x: x[:2]=='E0')]
    for year in years:
        data['Overall']['total'][year] = {}
        data['Overall']['total'][year]['value'] = round(topindex[topindex['AREANM'] == area].iloc[0][yearLU[year]], 1)
        data['Overall']['total'][year]['rank'] = list(topindex.sort_values(by=yearLU[year], ascending=False)['AREANM']).index(area)+1

        # CHANGE1YEAR
        if year == 2015:
            data[domain]['total'][year]['Change1year'] = 'NaN'
            data[domain]['total'][year]['Change1year Rank'] = 'NaN'
        else:
            change1 = topindex[topindex['AREANM'] == area].iloc[0][yearLU[year]] - topindex[topindex['AREANM'] == area].iloc[0][yearLU[year-1]]
            data[domain]['total'][year]['Change1year'] = round(change1, 1)

            change1list = sorted([topindex.loc[i][yearLU[year]] - topindex.loc[i][yearLU[year-1]] for i in topindex.index], reverse=True)
            data[domain]['total'][year]['Change1year Rank'] = change1list.index(change1) + 1

        # CHANGE4YEAR
        if year == 2019:
            change4 = topindex[topindex['AREANM'] == area].iloc[0][yearLU[2019]] - topindex[topindex['AREANM'] == area].iloc[0][yearLU[2015]]
            data[domain]['total'][year]['Change4year'] = round(change4, 1)

            change4list = sorted([topindex.loc[i][yearLU[2019]] - topindex.loc[i][yearLU[2015]] for i in topindex.index], reverse=True)
            data[domain]['total'][year]['Change4year Rank'] = change4list.index(change4) + 1
        else:
            data[domain]['total'][year]['Change4year'] = 'NaN'
            data[domain]['total'][year]['Change4year Rank'] = 'NaN'

    place = {}
    place["area"] = [topindex.iloc[i]['AREACD'] for i in topindex.index if topindex.iloc[i]['AREANM'] == area][0]
    place["name"] = area
    place["data"] = data
    # place["priority2018"]

    healthindexdata[place['area']] = place
```

```python
with open('healthindexdata.json', 'w') as outfile:
    json.dump(healthindexdata, outfile)
```
